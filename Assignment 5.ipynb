```python
import numpy as np
from math import comb
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def DAGcounter(n):
    
    if n == 0:
        return 1
    
    return n*(2**(n-1))*DAGcounter(n-1)
        
        
for i in range(10):
    print(DAGcounter(i))


```

```python
def DAGcount(n):
    
    if n == 0:
        return 1
    
    sum = 0
    
    for k in range(1,n+1):
        
        sum+= (-1)**(k+1) * comb(n,k) *(2**(k*(n-k)))*DAGcount(n-k)    
    
    
    
    return sum
        
        
for i in range(10):
    print(DAGcount(i))


```

```python

def make_spiral(arr):
    
    nrows, ncols= arr.shape
    idx = np.arange(nrows*ncols).reshape(nrows,ncols)[::-1]
    spiral_idx = []
    
    while idx.size:
        spiral_idx.append(idx[0])
        idx = idx[1:]
        idx = idx.T[::-1]

    spiral_idx = np.hstack(spiral_idx)
    spiral = np.empty_like(arr)
    spiral.flat[spiral_idx] = arr.flat[::-1]
    
    return spiral



for w in [3,5,10,100,1000]:
    primes = np.array([n for n in range(2,w**2+1) if all(
                            (n % m) != 0 for m in range(2,int(np.sqrt(n))+1))])

    arr = np.zeros(w**2, dtype='u1')
    arr[primes-1] = 1
    arr = make_spiral(arr.reshape((w,w)))
    plt.matshow(arr, cmap=cm.binary)
    plt.axis('off')
    plt.show()
```
