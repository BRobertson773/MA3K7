```python
n = 100

def sim(a = n,b = n, c= n):
    A, A_data = np.zeros(a), [a]
    B, B_data = np.zeros(b), [b]
    C, C_data = np.zeros(c), [c]
    
    t = 0
    time =[0]
    
    while (len(A) >=1 and len(B)>=1) or (len(A)>=1 and len(C)>=1) or (len(B)>=1 and len(C)>=1):
        t+=1
        time.append(t)
        
        tot = len(A)+len(B)+len(C)
        m1= random.choices([1,2,3],weights = [len(A)/tot,len(B)/tot,len(C)/tot],k=1)[0]
        
        if m1==1:
            A = np.delete(A,0)
        elif m1==2:
            B = np.delete(B,0)
        elif m1==3:
            C = np.delete(C,0)
    
        tot = len(A)+len(B)+len(C)
        m2= random.choices([1,2,3],weights = [len(A)/tot,len(B)/tot,len(C)/tot],k=1)[0]
        
        if m1 == m2:
            if m1==1:
                A = np.append(A,0)
            elif m1==2:
                B = np.append(B,0)
            elif m1==3:
                C = np.append(C,0)
                        
            A_data.append(len(A))
            B_data.append(len(B))
            C_data.append(len(C))      
            continue
                
        if m2==1:
            A = np.delete(A,0)
        elif m2==2:
            B = np.delete(B,0)
        elif m2 ==3:
            C = np.delete(C,0)
         
        if m1+m2==3:
            C = np.append(C,0)
        elif m1+m2 ==4:
            B = np.append(B,0)
        elif m1+m2 ==5:
            A = np.append(A,0)
        
        A_data.append(len(A))
        B_data.append(len(B))
        C_data.append(len(C))
    
    return time,A_data,B_data,C_data
```

```python
def count(particles):
    N = [0,0,0]
    
    for i in particles:
        if i[0] == 'A':
            N[0] +=1
        if i[0] == 'B':
            N[1] +=1
        if i[0] == 'C':
            N[2] +=1
    
    return N
```
