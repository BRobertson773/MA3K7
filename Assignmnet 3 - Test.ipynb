```python
import math
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
from collections import Counter 
import seaborn as sb
```

```python

def ascend(N):
    
    a_n = 0
    
    for n in range(1,N+1):
        a_n = n - a_n
    
    return a_n

print(ascend(2024))

def descend(N):
    
    a_n = N
    
    for n in range(1,N+1):
        a_n = abs((2024-n) - a_n)
    
    return a_n

print(descend(2024))
```

```python
def hat_number(N,length=2024):
    
    totals = []
    
    
    
    for n in range(1,N+1):
        
        hat = np.arange(1,length+1)
        hat = hat.tolist()
        for i in range(1,len(hat)):
            x = rnd.randint(0,len(hat)-1)
            p1 = hat[x]
            hat.pop(x)
            y = rnd.randint(0,len(hat)-1)
            p2 = hat[y]
            hat.pop(y)

            pnew = abs(p1-p2)

            hat.append(pnew)
        
        totals.append(hat[0])
        #if hat[0]%2 == 1:
        #        print('odd')
    
    return totals
```

```python
    
hat_number(1)

x = hat_number(10000)

y = Counter(x)



plt.bar(y.keys(),y.values())
plt.show()

sb.displot(y ,kde = True)
plt.show()

x.sort()
plt.plot(np.arange(1,len(x)+1),x)
plt.show()
print(max(x))
```
